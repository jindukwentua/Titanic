```python
plt.figure(figsize = (30,10))
sns.boxplot(x = 'Sex', y = 'Age', data = titanic)
```

```python
def Fill_age(data):
    age = data[0]
    sex = [1]
    
    
    if pd.isnull(age):
        if sex is 'male':
            return 29
        else:
            return 25
    else:
            return age
```

```python
 titanic['Age'] = titanic[['Age', 'Sex']].apply(Fill_age, axis = 1)
```

```python
 sns.heatmap(titanic.isnull(), yticklabels = False, cbar = False, cmap = 'Dark2')
```

```python
  titanic['Age'].hist( bins = 20, color = 'green')
```

```python
titanic.drop(['PassengerId'], axis = 1, inplace = True)
```

```python
male = pd.get_dummies(titanic['Sex'], drop_first = True)
```

```python
male
```

```python
titanic.drop(['Sex'], axis = 1, inplace = True)
```

```python
titanic
```

```python
titanic = pd.concat([titanic, male], axis = 1)
```

# MODEL TRAINING

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10)
```

```python
from  sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_predict = classifier.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot = True, fmt = 'd')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict))
```

```python
classifier.score(X_test, y_test)
```
